```python
#if data is not clean then step
data2=pd.read_csv("/home/aakash/Downloads/visualizing_data_with_d3/public/athlete_events.csv")
```

```python
# Remove null values and unneeded features
reviews = data2.dropna()
#reviews = reviews.drop(['Id','ProductId','UserId','ProfileName','HelpfulnessNumerator','HelpfulnessDenominator',   'Score','Time'], 1)
reviews = reviews.reset_index(drop=True)
```

```python
reviews.head()
```

```python
reviews.isnull().sum()
```

```python
print(len(reviews))
```

```python
#extra feature can add if required to count categorical data
reviews1 = pd.DataFrame({'count': [1 for i in range(len(reviews))]})
```

```python
reviews1.head()
```

```python
reviews1 = pd.concat([reviews1, reviews], axis=1) # reviews1.join(reviews)  can also use join   axis 1 is along row 
```

```python
cleanup_nums = {"Sex":     {"M": 1, "F": 0}}
reviews1.replace(cleanup_nums, inplace=True)
reviews1.head()
```

```python
cleanup_nums = {"Season":     {"Winter": 0, "Summer": 1}}
reviews1.replace(cleanup_nums, inplace=True)
reviews1.head()
```

```python
reviews1.to_csv("/home/aakash/Downloads/visualizing_data_with_d3/public/athlete_events_cleaned.csv")
```
